```python
teams = pd.DataFrame(columns=["pk1","pk2","pk3","pk4","pk5","Combined_score_1","Combined_score_2"])
```

```python
col_temp = pd.DataFrame(set(df["type1"]))
for i in range(100000):
    samp_temp = col_temp.sample(5,replace=False)[0]
    temp1 = df[df["type1"].isin([samp_temp.values[0]])].sample(1)
    temp2 = df[df["type1"].isin([samp_temp.values[1]])].sample(1)
    temp3 = df[df["type1"].isin([samp_temp.values[2]])].sample(1)
    temp4 = df[df["type1"].isin([samp_temp.values[3]])].sample(1)
    temp5 = df[df["type1"].isin([samp_temp.values[4]])].sample(1)
    temp = pd.concat([temp1,temp2,temp3,temp4,temp5])
    list1 = temp["pokedex_number"].to_list()
    list1.append(temp["Combined_score_1"].sum())
    list1.append(temp["Combined_score_2"].sum())
    teams = teams.append(pd.DataFrame([list1],columns=["pk1","pk2","pk3","pk4","pk5","Combined_score_1","Combined_score_2"]))
```

```python
teams.sort_values(by = ["Combined_score_2","Combined_score_1"],ascending=[False,True],inplace=True)
```

```python
teams.head()
```

```python
(df.loc[df["pokedex_number"].isin(teams.iloc[0,0:5])])[["name","type1","type2","Combined_score_1","Combined_score_2"]]
```

```python
import pickle

# Save to file in the current working directory
pkl_filename1 = "pickle_teams.pkl"
with open(pkl_filename1, 'wb') as file:
    pickle.dump(teams, file)
```

```python
# Load from file
with open(pkl_filename1, 'rb') as file:
    teams = pickle.load(file)
```
